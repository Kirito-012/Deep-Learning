```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
f(3.0)
```

```python
xs = np.arange(-5,5,0.25)
ys = f(xs)
plt.plot(xs, ys)
plt.xlabel('xs')
plt.ylabel('ys')
plt.show()
```

```python
x = -3
h = 0.001
# y = f(x+h) 
y = (f(x+h) - f(x))/h
print(y)


plt.plot(xs,ys)
plt.scatter(x, f(x+h))
plt.show()
```
